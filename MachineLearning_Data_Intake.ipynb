#Train/ test split

##Balancing around S02 daily max >15

```python
#In order to run logistic regression and other classification algorithms we need to create a binary variable
#corresponding to dangerous S02, ie y = 1 if AQI>10 and 0 otherwise
#10 is an important threshold because according to the CDC that is around
#when certain groups start experiencing adverse health effects due to poor air quality in less than 1 hour of exposure

corr_df = ML_df.drop(columns = [ 'State', 'Year', 'Date', 'Address', 'Event Type', 'City Name', 'S02 Daily Mean'])

S02_series = list(corr_df['S02 Daily Max'])
bad_S02 = []
for i in range(len(S02_series)):
    
    if(S02_series[i]>15):
        y=1
        bad_S02.append(y)
    else:
        y=0
        bad_S02.append(y)
col_values = pd.Series(bad_S02)
corr_df.insert(loc =2 , column = 'S02>15', value= col_values)


#fixing class imbalance


#before splitting into train and test sets, let's downsample the negative examples to acheive class balance
#otherwise the classifier could do quite well just by predicting '0' indiscriminately
#to do this we will resample from the majority class without replacement to get the same #samples as the minority
df_majority = corr_df[corr_df['S02>15']==0]
df_minority = corr_df[corr_df['S02>15']==1]
 
# Downsample majority class
df_majority_downsampled = resample(df_majority, 
                                 replace=False,    # sample without replacement
                                 n_samples=df_minority.shape[0],     # to match minority class
                                 random_state=123) # reproducible results
 
# Combine minority class with downsampled majority class
balancedS02_df = pd.concat([df_majority_downsampled, df_minority])

balancedS02_df.reset_index(inplace=True)
balancedS02_df.drop(columns = ['index'], inplace=True)
#balanced_df
 
# Display new class counts
balancedS02_df['S02>15'].value_counts()
```

##Balancing around AQI>50
